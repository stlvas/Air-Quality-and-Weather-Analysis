## Keeping only California's- Los Angeles data

```python
# Also aggregating (avg max temperatures to identify heatwaves)
```

```python
los_angeles_data = california[california['County Name'] == 'Los Angeles'].copy()

# Calculating average temperature and max_value for all sites within Los Angeles County
los_angeles_avg = los_angeles_data.groupby(['Date Local']).agg({
    'Temperature(C) Max Value': 'mean',
    'max_value': 'mean'
}).reset_index()
```

## Identifying heatwaves in Los Angeles

```python
def identify_heatwaves2(df2, threshold_temp=34.0):
    heatwave_events2 = []
    df_sorted2 = df2.sort_values(by='Date Local').reset_index(drop=True)
    
    i = 0
    while i < len(df_sorted2):
        # Check if current day exceeds threshold temperature
        if df_sorted2.loc[i, 'Temperature(C) Max Value'] > threshold_temp:
            start_date2 = df_sorted2.loc[i, 'Date Local']
            # Initialize end date to start date
            end_date2 = start_date2
            # Move to next day
            i += 1
            # Check subsequent days as long as temperatures are above threshold
            while (i < len(df_sorted2)) and (df_sorted2.loc[i, 'Temperature(C) Max Value'] > threshold_temp):
                end_date2 = df_sorted2.loc[i, 'Date Local']
                i += 1
            # Add heatwave event to list
            heatwave_events2.append({
                'Start Date': start_date2,
                'End Date': end_date2
            })
        else:
            i += 1
    
    return pd.DataFrame(heatwave_events2)


heatwave_events2 = identify_heatwaves2(los_angeles_avg)

heatwave_events2
```

```python
heatwave_events_filtered = heatwave_events2[(heatwave_events2['End Date'] - heatwave_events2['Start Date']).dt.days >= 2]
heatwave_events_filtered
```

## Creating plots for  heatwaves in 2020

```python
data_2020 = los_angeles_avg[los_angeles_avg['Date Local'].dt.year == 2020]

# Extract month and day from 'Date Local'
data_2020['Month'] = data_2020['Date Local'].dt.month
data_2020['Day'] = data_2020['Date Local'].dt.day

# Create a list of months for labeling purposes
month_names = ['January', 'February', 'March', 'April', 'May', 'June', 
               'July', 'August', 'September', 'October', 'November', 'December']

# Iterate over each month in 2020
for month in sorted(data_2020['Month'].unique()):
    # Filter data for the current month
    month_data = data_2020[data_2020['Month'] == month]
    
    # Skip if no data for this month
    if month_data.empty:
        continue
    
    # Create a figure and axis for plotting
    fig, ax1 = plt.subplots(figsize=(12, 6))
    
    # Plot AQI on primary y-axis (left)
    ax1.plot(month_data['Day'], month_data['max_value'], marker='o', linestyle='-', color='#004753', label='AQI')
    ax1.set_xlabel('Day', fontsize=18)
    ax1.set_ylabel('AQI', color='#004753', fontsize=18)
    ax1.tick_params(axis='y', labelcolor='#004753')
    
    # Create a secondary y-axis for Temperature (right)
    ax2 = ax1.twinx()
    ax2.plot(month_data['Day'], month_data['Temperature(C) Max Value'], marker='o', linestyle='-', color='#A3D0D4', label='Temperature(C) Max Value')
    ax2.set_ylabel('Temperature(C) Max Value', color='#A3D0D4', fontsize=18)
    ax2.tick_params(axis='y', labelcolor='#A3D0D4')
    
    # Highlight heatwave events within this month
    for index, row in heatwave_events_filtered.iterrows():
        # Check if the heatwave event starts or ends in the current month
        if (row['Start Date'].year == 2020 and row['Start Date'].month == month) or \
           (row['End Date'].year == 2020 and row['End Date'].month == month):
            
            # Determine the start and end days for highlighting
            start_day = row['Start Date'].day if row['Start Date'].month == month else 1
            end_day = row['End Date'].day if row['End Date'].month == month else month_data['Day'].max()
            
            # Highlight the heatwave event
            ax1.axvspan(start_day, end_day, color='#CCAA14', alpha=0.3)
    
    # Title and layout
    plt.title(f'AQI and Temperature(C) per Day in {month_names[month-1]} 2020 with Heatwave Events', fontsize=20)
    
    # Grid and layout
    ax1.grid(True)
    plt.tight_layout()
    plt.show()
```
