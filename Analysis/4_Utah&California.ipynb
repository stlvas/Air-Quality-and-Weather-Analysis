```python
import pandas as pd
```

## Selecting California and Utah for air quality and AQI data

```python
#Import daily air quality data
param_all = pd.read_csv("air_data_merged.csv")
```

```python
param_all.head()
```

```python
# Keeping only Utah and California
filtered_df = param_all[param_all['State Name'].isin(['Utah', 'California'])]
filtered_df
```

```python
filtered_df.info()
```

```python
param_all[param_all['State Code'].isin([6, 49])].isna().sum()
```

```python
#Changing data types dtypes
```

```python
filtered_df['State Code'] = filtered_df['State Code'].astype(int).astype(str)
```

```python
filtered_df['County Code'] = filtered_df['County Code'].astype(int).astype(str)
```

```python
filtered_df['Site Num'] = filtered_df['Site Num'].astype(int).astype(str)
```

```python
filtered_df['Date Local'] = pd.to_datetime(filtered_df['Date Local'])
```

```python
filtered_df.info()
```

```python
#Dropping unnecessary columns
filtered_df = filtered_df.drop(columns=['AQI', 'CBSA Name'])
filtered_df
```

```python
# Importing daily AQI csv
aqi = pd.read_csv("daily_AQI.csv")
aqi.head()
```

```python
aqi.info()
```

```python
#Keeping data only from California and Utah
aqi_utah_cal = aqi[aqi['State Name'].isin(['Utah', 'California'])]
aqi_utah_cal
```

```python
#Merging AQI data for Utah and California in the df
merg = pd.merge(filtered_df, aqi_utah_cal, on='measurement_id', how='outer')
merg
```

```python
merg.isna().sum()
```

```python
merg_cols= merg.columns
```

```python
#Dropping unnecessary columns
merg= merg.drop(columns=merg.filter(regex='_y$').columns)
merg
```

```python
#Renaming Columns
merg.rename(columns={'State Code_x':'State Code', 'State Name_x':'State Name', 'County Code_x':'County Code', 'County Name_x':'County Name',
       'City Name_x':'City Name', 'Site Num_x':'Site Num', 'Datum_x':'Datum', 'Latitude_x':'Latitude', 'Longitude_x':'Longitude',
       'Date Local_x':'Date Local'},inplace=True)
```
