```python
import pandas as pd
```

## Selecting weather data only from California and Utah

```python
#Import daily weather data
weather_df = pd.read_csv("weather_data.csv")
```

```python
#Slice data for California and Utah
weather_utah_cali = weather_df[(weather_df['State Name']=='Utah')|(weather_df['State Name']=='California')]
```

```python
#Save to csv
weather_utah_cali.to_csv("utah_cali_daily_weather.csv",index=False)
```

## Selecting air quality and AQI data only from California and Utah

```python
#Import daily air quality data
param_all = pd.read_csv("air_data_merged.csv", low_memory=False)
```

```python
# Keeping only Utah and California
filtered_df = param_all[param_all['State Name'].isin(['Utah', 'California'])]
filtered_df
```

```python
#Changing data types dtypes
filtered_df['State Code'] = filtered_df['State Code'].astype(int).astype(str)
filtered_df['County Code'] = filtered_df['County Code'].astype(int).astype(str)
filtered_df['Site Num'] = filtered_df['Site Num'].astype(int).astype(str)
filtered_df['Date Local'] = pd.to_datetime(filtered_df['Date Local'])
```

```python
#Dropping unnecessary columns
filtered_df = filtered_df.drop(columns=['AQI', 'CBSA Name'])
filtered_df
```

```python
# Importing daily AQI csv
aqi = pd.read_csv("daily_AQI.csv")
aqi.head()
```

```python
#Keeping data only from California and Utah
aqi_utah_cal = aqi[aqi['State Name'].isin(['Utah', 'California'])]
aqi_utah_cal
```

```python
#Merging AQI data for Utah and California in the df
aqi_utah_cali = pd.merge(filtered_df, aqi_utah_cal, on='measurement_id', how='outer')
aqi_utah_cali
```

```python
merg_cols= aqi_utah_cali.columns
```

```python
#Dropping unnecessary columns
aqi_utah_cali= aqi_utah_cali.drop(columns=aqi_utah_cali.filter(regex='_y$').columns)
aqi_utah_cali
```

```python
#Renaming Columns
aqi_utah_cali.rename(columns={'State Code_x':'State Code', 'State Name_x':'State Name', 'County Code_x':'County Code', 'County Name_x':'County Name',
       'City Name_x':'City Name', 'Site Num_x':'Site Num', 'Datum_x':'Datum', 'Latitude_x':'Latitude', 'Longitude_x':'Longitude',
       'Date Local_x':'Date Local'},inplace=True)
```

```python
#Saving the final df for Utan and California to csv
aqi_utah_cali.to_csv("daily_Cal&Utah.csv", index=False, encoding='utf-8')
```

## Merging air quality and AQI with weather data

```python
#Merge weather with air quality and AQI data
all_df = pd.merge(aqi_utah_cali, weather_utah_cali, how='outer',on='measurement_id')
```

```python
#Fill left columns if NaN
for column in all_df.columns:
    if column.endswith('_x'):
        corresponding_column = column[:-2] + '_y'
        if corresponding_column in all_df.columns:
            all_df[column] = all_df[column].fillna(all_df[corresponding_column])

#Drop right columns
all_df = all_df.drop(columns=all_df.filter(regex='_y$').columns)

#Rename columns
all_df.rename(columns={'State Code_x':'State Code', 'State Name_x':'State Name', 'County Code_x':'County Code', 'County Name_x':'County Name',
       'City Name_x':'City Name', 'Site Num_x':'Site Num', 'Datum_x':'Datum', 'Latitude_x':'Latitude', 'Longitude_x':'Longitude',
       'Date Local_x':'Date Local','CBSA Name_x':'CBSA Name'},inplace=True)
```

```python
# Save to CSV
all_df.to_csv("utah_cali_all.csv",index=False)
```
